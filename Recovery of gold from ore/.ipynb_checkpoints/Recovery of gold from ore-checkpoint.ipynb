```python
STAGE = ['rougher.input.feed', 'rougher.output.concentrate', 'primary_cleaner.output.concentrate', 'final.output.concentrate']

for step in STAGE:
    data_train = data_train[(data_train[f'{step}_au'] != 0) &
                           (data_train[f'{step}_ag'] != 0) &
                           (data_train[f'{step}_pb'] != 0)]
```

## Модель

### Функции для расчета sMAPE и итоговой sMAPE

```python
def sMAPE (target, predictions):
    absolute = abs(target - predictions) / ((abs(target) + abs(predictions)) / 2)
    summa = absolute[:len(target)].sum()
    smape = (1 / len(target)) * summa * 100
    return smape * -1
```

```python
scorer_smape = make_scorer(sMAPE, greater_is_better=False)
```

```python
def final_sMAPE (smape_rougher, smape_final):
    final = 0.25 * smape_rougher + 0.75 * smape_final
    return final
```

### Разделение на выборки

```python
features_train = data_train.drop(data_full.columns.difference(data_test.columns), axis=1).reset_index(drop=True)
features_train = features_train.drop(['recovery'], axis=1).reset_index(drop=True)

target_train_rougher = data_train['rougher.output.recovery'].reset_index(drop=True)
target_train_final = data_train['final.output.recovery'].reset_index(drop=True)

features_test_rougher = data_test.reset_index(drop=True)
target_test_rougher = pd.Series(data_full['rougher.output.recovery'], index = data_test.index).reset_index(drop=True)

features_test_final = data_test.reset_index(drop=True)
target_test_final = pd.Series(data_full['final.output.recovery'], index = data_test.index).reset_index(drop=True)
```

```python
print(features_train.shape, target_train_rougher.shape, features_train.shape, target_train_final.shape,
features_test_rougher.shape, target_test_rougher.shape, features_test_final.shape, target_test_final.shape)
```

### Выбор модели на тест

#### Модель Дерево решений

```python
%%time
best_model = None
best_result = 20
for depth in range(1, 100, 5):
    model_r = DecisionTreeRegressor(random_state=12345, criterion='friedman_mse', max_depth=depth, splitter='random')
    result = cross_val_score(model_r, features_train, target_train_rougher, cv=5, scoring = scorer_smape).mean()
    if result < best_result:
        best_model = model_r
        best_result = result
print(f'Симметричное среднее абсолютное процентное отклонение: {best_result:.3f}; Глубина дерева: {depth}')
```

```python
%%time
best_model_f = None
best_result_f = 20
for depth_f in range(1, 100, 5):
    model_f = DecisionTreeRegressor(random_state=12345, max_depth=depth)
    result_f = cross_val_score(model_f, features_train, target_train_final, cv=5, scoring = scorer_smape).mean()
    if result_f < best_result_f:
        best_model_f = model_f
        best_result_f = result_f
print(f'Симметричное среднее абсолютное процентное отклонение: {best_result_f:.2f}; Глубина дерева: {depth_f}')
```

#### Модель Случайный лес

```python
%%time
best_model = None
best_result = 20
for est in range(10, 50, 10):
    for depth in range(1, 13, 2):
        model_r = RandomForestRegressor(n_estimators=est, max_depth=depth, random_state=12345, criterion='friedman_mse')
        result = cross_val_score(model_r, features_train, target_train_rougher, cv=5, scoring = scorer_smape, error_score='raise').mean()
        if result < best_result:
            best_model = model_r
            best_result = result
print(f'Симметричное среднее абсолютное процентное отклонение: {best_result:.2f}; Кол-во деревьев: {est}; Глубина дерева: {depth}')
```

```python
%%time
best_model = None
best_result = 20
for est in range(10, 50, 10):
    for depth in range(1, 13, 2):
        model_f = RandomForestRegressor(n_estimators=est, max_depth=depth, random_state=12345, criterion='friedman_mse')
        result = cross_val_score(model_f, features_train, target_train_final, cv=5, scoring = scorer_smape).mean()
        if result < best_result:
            best_model = model_r
            best_result = result
print(f'Симметричное среднее абсолютное процентное отклонение: {best_result:.2f}; Кол-во деревьев: {est}; Глубина дерева: {depth}')
```

#### Модель Линейной регрессии

```python
model_r =  LinearRegression()
model_r.fit(features_train, target_train_rougher)
result = cross_val_score(model_r, features_train, target_train_rougher, cv=5, scoring = scorer_smape)
result = result.mean()

print(f'Симметричное среднее абсолютное процентное отклонение: {result:.2f}')
```

```python
model_f =  LinearRegression()
model_f.fit(features_train, target_train_rougher)
result = cross_val_score(model_f, features_train, target_train_final, cv=5, scoring = scorer_smape).mean()

print(f'Симметричное среднее абсолютное процентное отклонение: {result:.2f}')
```

### Вывод

Лучшей моделью можно считать Случайный лес с гиперпараметрами:

- показатель: 6.66

- глубина дерева: 11
- criterion: MSE

## Тестирование

```python
model_r = RandomForestRegressor(n_estimators=40, max_depth=11, random_state=12345, criterion='friedman_mse')
model_r.fit(features_train, target_train_rougher)
predicted_r = model_r.predict(features_test_rougher) 
smape_rougher = sMAPE(target_test_rougher, predicted_r) * -1


model_f = RandomForestRegressor(n_estimators=40, max_depth=11, random_state=12345, criterion='friedman_mse')
model_f.fit(features_train, target_train_final)
predicted_f = model_f.predict(features_test_final)
smape_final = sMAPE(target_test_final, predicted_f) * -1

print(f'''Симметричное среднее абсолютное процентное отклонение для черновой:{smape_rougher:.2f};
Симметричное среднее абсолютное процентное отклонение для финальной:{smape_final:.2f};
Итоговое симметричное среднее абсолютное процентное отклонение:{final_sMAPE(smape_rougher, smape_final):.2f}''')
```

```python
dummy_regr_r = DummyRegressor(strategy='median')
dummy_regr_r.fit(features_train, target_train_rougher)
dummy_predicted_r = dummy_regr_r.predict(features_test_rougher)
dummy_smape_rougher = sMAPE(target_test_rougher, dummy_predicted_r) * -1

dummy_regr_f = DummyRegressor(strategy='median')
dummy_regr_f.fit(features_train, target_train_final)
dummy_predicted_f = dummy_regr_f.predict(features_test_final)
dummy_smape_final = sMAPE(target_test_final, dummy_predicted_f) * -1

print(f'''Константная модель:
Симметричное среднее абсолютное процентное отклонение для черновой:{dummy_smape_rougher:.2f};
Симметричное среднее абсолютное процентное отклонение для финальной:{dummy_smape_final:.2f};
Итоговое симметричное среднее абсолютное процентное отклонение:{final_sMAPE(dummy_smape_rougher, dummy_smape_final):.2f}''')
```
